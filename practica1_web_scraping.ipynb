```python
import requests
import re
from bs4 import BeautifulSoup

class CompetitorScraper():
    
    def __init__(self):
        self.url = "https://muebleslufe.com"

    def download(self, url, num_retries=2, user_agent='wswp'):
        print('Downloading:', url)
        headers = {'User-agent': 'wswp'}
        try:
            html = requests.get(url, headers=headers)
        except (URLError, HTTPError, ContentTooShortError) as e:
            print('Download error:', e.reason)
            html = None
            if num_retries > 0:
                if hasattr(e, 'code') and 500 <= e.code < 600:
                    return download(url, num_retries - 1)                
        return html.content

    def get_category_links(self):
        url = self.download(self.url)
        soup = BeautifulSoup(url)
        a_tags = soup.body.find_all('a')
        pattern=re.compile('{0}/es/[0-9]+-'.format(self.url))
        return {link.get('href') for link in a_tags if re.match(pattern, str(link.get('href')))}
    
    def get_items_links(self, category_page):
        url = self.download(category_page)
        soup = BeautifulSoup(url)
        a_tags = soup.body.find_all('a')
        pattern=re.compile('{0}/es/[\w-]+/[0-9]+-[\w-]*.html'.format(self.url))
        return {link.get('href') for link in a_tags if re.match(pattern, str(link.get('href')))}
    
    def get_features(self, item_page):
        pass
    
    def data2csv(self, filename):
        # Overwrite to the specified file.
        # Create it if it does not exist.
        file = open("../csv/" + filename, "w+")

        # Dump all the data with CSV format
        for i in range(len(self.data)):
            for j in range(len(self.data[i])):
                file.write(self.data[i][j] + ";");
            file.write("\n");

            
output_file = "dataset.csv"
scraper = CompetitorScraper()
cat_links = scraper.get_category_links()
cat_links


items_links = scraper.get_items_links('https://muebleslufe.com/es/10-matrimonio')
items_links


url = scraper.download('https://muebleslufe.com/es/camas/209-cama-de-matrimonio-con-lamas.html')
soup = BeautifulSoup(url)


#scraper.scrape();
#scraper.data2csv(output_file);


```
